```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 선그래프
x = [1,2,3]
y = [5,7,4]
plt.plot(x, y)
plt.show()
```

```python
# 선그래프
x = [1,2,3]
y1 = [5,7,4]
y2 = [10,15,6]
plt.plot(x, y1, label='line1')
plt.plot(x, y2, label='line2')
plt.title('Line Chart')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

```python
# 산점도
import numpy as np
x = np.arange(1,9)
y = [3,5,6,8,3,8,9,3]
plt.scatter(x,y)
plt.title('Scatter Plot')
plt.show()
```

```python
x = np.arange(1,9)
y = [3,5,6,8,3,8,9,3]
plt.scatter(x,y, marker='x', color='yellow', s=100)
plt.title('Scatter Plot\n Example')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(0,10)
plt.xlim(0,10)
plt.show()
```

```python
data = [[2, 4, 6, 8, 10],[6,7, 8, 2, 4],[1, 3, 5, 7, 9],[7, 8, 2, 4, 2]]
x = pd.DataFrame(data)
plt.boxplot(x)
plt.show()
```

```python
days = [1, 2, 3, 4, 5]
sleeping = [7, 8, 6, 11, 7]
eating = [2, 3, 4, 3, 2]
working = [7, 8, 7, 2, 2]
playing = [8, 5, 7, 8, 13]
plt.stackplot(days, sleeping, eating, working, playing)
plt.legend(labels=['sleeping', 'eating', 'working', 'playing'])
plt.show()
```

```python
data=[[2,3,4,1],[6,3,5,2],[6,3,5, 4],[3,7,5,4],[2,8,1,5]] #5x4 matrix
Index= ['I1', 'I2','I3','I4','I5']
Cols = ['C1', 'C2', 'C3','C4'] #4
df = pd.DataFrame(data, index=Index, columns=Cols)
plt.pcolor(df.corr())
#plt.matshow(df.corr())
plt.colorbar() 
plt.show()
```

# Analyze the Chipotle data using graphs 
## Scatter plot
## Histogram
## Bar chart, horizontal bar chart
## Heat map

```python
# price 와 quantity 상관관계

dir = "./assets/chipotle.xlsx"
chipo = pd.read_excel(dir)
chipo
```

```python
x = chipo.quantity
x
```

```python
y = chipo.item_price
y
```

```python
plt.title('scatter plot of \n item price and quantity')
plt.scatter(x,y)
plt.xlabel('chipo qunatity')
plt.ylabel('item price')
plt.show()
```

```python
plt.title('bar chart of \n item price and quantity')
plt.xlabel('chipo qunatity')
plt.ylabel('item price')
plt.bar(x, y)
plt.show()
```

```python
plt.hist(y, bins=30)
plt.title('Price Histogram')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

```python
# Bar chart, horizontal bar chart
#chipo.groupby('item_name').mean().item_price
tab = chipo.item_name.value_counts()
plt.barh(tab.index, tab)
plt.show()
```

```python
# Heat map
corr = chipo[['quantity', 'item_price']].corr()
plt.matshow(corr)
plt.colorbar()
plt.title('Heat map')
plt.show()
```
