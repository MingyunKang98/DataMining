```python
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
from sklearn.linear_model import SGDRegressor,GammaRegressor,Lasso,GammaRegressor,ElasticNet,Ridge
from sklearn.linear_model import RANSACRegressor,HuberRegressor, BayesianRidge,LinearRegression
from sklearn.ensemble import RandomForestRegressor, BaggingRegressor, AdaBoostRegressor, GradientBoostingRegressor, ExtraTreesRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor # Decision Tree Regression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import Pipeline # Streaming pipelines
from sklearn.model_selection import learning_curve, validation_curve, GridSearchCV # Model evaluation
from sklearn.preprocessing import StandardScaler
import statsmodels.api as sm
```

```python
data = df.copy()
```

```python
X=data.drop(columns='C_str')
Y=data['C_str']
```

```python
X_ = sm.add_constant(X)
model1 = sm.OLS(Y,X_).fit()
model1.summary()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.25, shuffle=True)
```

```python
sc=StandardScaler()
X_train=sc.fit_transform(X_train) #each value is now its z-score
X_test=sc.transform(X_test)
```

```python
lr=LinearRegression()
sgd=SGDRegressor()
lasso=Lasso()
ridge=Ridge()
rf=RandomForestRegressor()
dt=DecisionTreeRegressor()
gboost=GradientBoostingRegressor()
bagging=BaggingRegressor()
adboost=AdaBoostRegressor()
knn=KNeighborsRegressor()
```

```python
algo=[lr,sgd,lasso,ridge,rf,dt,gboost,bagging,adboost,knn]
```

```python
models=[]
accuracy_test=[]
accuracy_train=[]
for i in range(len(algo)):
    algo[i].fit(X_train,y_train)
    accuracy_train.append(algo[i].score(X_train,y_train))
    accuracy_test.append(algo[i].score(X_test,y_test))
    models.append(algo[i])
```

```python
import torch
import torch.optim as optim
import torch.nn as nn
```

```python
X_train=torch.Tensor(X_train)
y_train=torch.Tensor(y_train)
X_test=torch.Tensor(X_test)
y_test = torch.from_numpy(y_test.to_numpy())
y_test = torch.as_tensor(y_test)
```

```python
class ANN(nn.Module):
  def __init__(self, input_size,hl1,hl2,output_size):
    super().__init__()
    self.l1=nn.Sequential(nn.Linear(input_size,hl1,bias=True), nn.ReLU())
    self.l2=nn.Sequential(nn.Linear(hl1,hl2,bias=True), nn.Sigmoid()) #linear layer with a relu activation function (output size = hl1)
    self.output=nn.Sequential(nn.Linear(hl2,output_size,bias=True)) #only linear layer with required output size i.e. 1 (for our case)

  def forward(self, x): #feeding forward
    data=x
    data=self.l1(data)
    data=self.l2(data)
    data=self.output(data)
    return data
```

```python
input_size=6
hs1=24
hs2=24
output=1
model=ANN(input_size,hs1,hs2,output)
```

```python
criterion=nn.MSELoss() #MSELoss is the loss function (Mean Squared Error)
optimizer=optim.AdamW(model.parameters(), lr=0.01)
```

```python
from sklearn.metrics import r2_score
```

```python
epochs=10000
losses=[]
acc=[]
for i in range(epochs):
  optimizer.zero_grad()
  outputs=model(X_train).squeeze()
  loss=criterion(outputs,y_train) #calculate loss
  loss.backward() #backpropagation
  optimizer.step()

  losses.append(loss.item())
  acc.append(r2_score(outputs.detach().numpy(),y_train.detach().numpy()))

  print('Epoch [{}/{}]'.format(i+1,epochs),' r2: {:.4f}'.format(acc[-1]))
```

```python
X_test=torch.Tensor(X_test)
y_test=torch.Tensor(y_test)
with torch.no_grad():
  predicted=model(X_test).squeeze()
  r2=r2_score(predicted.detach().numpy(), y_test.detach().numpy())
  print('Accuracy Score : {:4f}%'.format(r2*100))
```

```python
models.append('ANN')
accuracy_train.append(acc[-1])
accuracy_test.append(r2)
```

```python
mod=pd.DataFrame([models,accuracy_train,accuracy_test]).T
mod.columns=['model','score_train','score_test']
mod
```

```python
mod.score_test = mod.score_test.astype(float)
plt.figure(figsize=(7,7))
plt.barh(range(11), mod.score_test)
plt.yticks(range(11), mod.model)
plt.xlim(0.5,1)
for index, value in enumerate(mod.score_test):
    plt.text(value, index, str(round(value*100,2))+'%')
plt.show()
```
