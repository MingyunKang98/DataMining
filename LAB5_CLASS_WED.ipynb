```python
x = np.arange(10)
y = x**2 #제곱
z = x**3 #3승
t = x**4 #4승
```

```python
plt.plot(x,y)
plt.plot(x,z)
plt.plot(x,t)
plt.show()
```

```python
# 선그래프
plt.plot(x,y, label="Race1")
plt.plot(x,z, label="Race2")
plt.plot(x,t, label="Race3")
plt.legend()
plt.show()
```

```python
# 범례 넣는 다른 방법
plt.plot(x,y)
plt.plot(x,z)
plt.plot(x,t)
plt.legend(labels=['Race1', 'Race2', 'Race3'], loc=5)
plt.show()
```

```python
lst = [1, np.nan, 'hello', None]
s = pd.Series(lst)
s.isna() #널값 체크
s.isnull() #널값 체크
s.isna().sum() #널값 갯수 확인
s[s.isna()] #널값 서브셋
s[s.notnull()] #널값이 아닌 데이터 서브셋
```

```python
s.dropna() #널값 드랍할 때
s
```

```python
# 데이터프레임에서 널값 처리
data = [[1, np.nan, 2], [2,3,5], [np.nan, 4, 6]]
df = pd.DataFrame(data)
df
```

```python
df.dropna() #널값이 있는 행 삭제
df.dropna(axis=1) #널값이 있는 열 삭제
```

```python
df = pd.DataFrame(data, columns=['a', 'b', 'c'])
```

```python
# 컬럼별로 널값 갯수 확인
df.isna().sum()
```

```python
# 전체 데이터에서 널값 갯수
df.isna().sum().sum()
```
